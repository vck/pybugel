```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer as tf
from sklearn.feature_extraction.text import CountVectorizer as cv
from sklearn.model_selection import train_test_split
```

```python
DATA_PATH = "../data/query.csv"

#load data into pandas dataframe

df = pd.read_csv(DATA_PATH)
```

```python
# preprocessing and cleaning 
# remove NaN and NA from dataframe

df.head()
```

```python
# first attempt, try to include all of the punctuation on the title as the valid datasets
# but beforehand, we need to remove the None and NaN val

df = pd.concat([df.title, df.type], axis=1, keys=["title", "type"])
```

```python
df = df.mask(df.eq('None')).dropna()
```
